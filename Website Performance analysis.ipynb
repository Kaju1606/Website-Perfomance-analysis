```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

Step 3:- First process is data cleaning here we remove upper row which is not relevent instead of that row we placed 0 row data as header.We need to reset the index starting from 0 and rename the columns into meaningful name.

```python
df.columns = df.iloc[0]
df = df.drop(index =0).reset_index(drop = True)
df.head()
```

```python
df.columns = ["Channel group", "Date hour", "Users", "Sessions", "Engaged sessions", "Average engagement time per sessions", "Engaged sessions per user", "Event per sessions", "Engagement Rate", "Event Count"]
```

Step 5:- Now we can able to see each column assign as object data type which is wrong user to last each column has store numerical type of data and Date hour column is also indicate DD-mm-yyyy formate so we change data type and create new column which is hour we need to seprated into new columns.

```python
df.head()
```

```python
df["Date hour"] = pd.to_datetime(df["Date hour"], format="%Y%m%d%H", errors ='coerce')
df.head()
```

```python
df.info()
```

```python
numeric_cols = df.columns.drop(["Channel group", "Date hour"])
df[numeric_cols]= df[numeric_cols].apply(pd.to_numeric, errors='coerce')
df["hour"] = df["Date hour"].dt.hour
```

Problem 5:- which channels are driving more engaged session compared to non engaged ones and what strategies can improve engagement in underperforming channels

```python
df.head()
```

```python
session_df = df.groupby("Channel group")[["Sessions", "Engaged sessions"]].sum().reset_index()
session_df["Non-Engaged"]= session_df["Sessions"] - session_df["Engaged sessions"]
session_df_melted = session_df.melt(id_vars="Channel group", value_vars=["Engaged sessions", "Non-Engaged"])

plt.figure(figsize=(8,5))
sns.barplot(data=session_df_melted, x="Channel group", y="value", hue="variable")
plt.title("Engaged vs Non Engaged sessions")
plt.xticks(rotation=45)
plt.show()
```

Conclusion is that organic social is a medium where user can visit on website enaged is more as compared to other.

Problem 6:- At what hours of the day does each channel drive the most traffic

```python
heatmap_data = df.groupby(["hour", "Channel group"])["Sessions"].sum().unstack().fillna(0)

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=.5, annot=True, fmt='.0f')
plt.title("Traffic by hour and channels")
plt.xlabel("Channel group")
plt.ylabel("Hour of the day")
plt.show()
```

Conclusion here we can able to see most traffic is on the organic social at 12:00 pm. Referral part most traffic in 11:00 am

Problem 7:- is there any correlation between high tarffic and high engagement rate over time?

```python
df.head()
    
```

```python
df_plot = df.groupby("Date hour")[["Engagement Rate", "Sessions"]].mean().reset_index()

plt.figure(figsize=(10,5))
plt.plot(df_plot["Date hour"], df_plot["Engagement Rate"], label="Engagement Rate", color= "green")
plt.plot(df_plot["Date hour"], df_plot["Sessions"], label="Sessions", color="blue")
plt.title("Engagement Rate Vs Sessions Over Time")
plt.xlabel("Date hour")
plt.legend()
plt.grid(True)
plt.show()
```

Conclusion is that the session line is up down but engagement rate is flat minimum
